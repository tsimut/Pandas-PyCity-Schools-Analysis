```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load 
school_info="../Downloads/schools_complete.csv"
student_info="../Downloads/students_complete.csv"
# Read School and Student Data File and store into Pandas Data Frames
school_list = pd.read_csv(school_info)
student_list = pd.read_csv(student_info)
# Combine the data into a single dataset
combined_lists = pd.merge(student_list, school_list, how="left", on=["school_name", "school_name"])
```

```python
#DISTRICT SUMMARY
#Calculate the total number of schools
total__schools = len(combined_lists["school_name"].unique())
#Calculate the total number of students
total_students = len(combined_lists["Student ID"])
#Calculate the total budget
total_budget = school_list["budget"].sum()
#Calculate the average math score
mean_math_score=combined_lists["math_score"].mean()
#Calculate the average reading score  
mean_reading_score=combined_lists["reading_score"].mean()
#Calculate the overall passing rate error
overall_passing_rate=(mean_math_score + mean_reading_score)/2
#Calculate the percentage of students with a passing math score
passing_math=combined_lists.loc[combined_lists["math_score"] >= 70]
percentage_passing_math=(len(passing_math["Student ID"])/total_students)*100
#Calculate the percentage of students with a passing reading score
passing_reading=combined_lists.loc[combined_lists["reading_score"] >= 70]
percentage_passing_reading=(len(passing_reading["Student ID"])/total_students)*100
```

```python
# Place all of the data found into a summary DataFrame
district_summary= pd.DataFrame({"Total Schools":[total__schools] ,
                              "Total Students": [total_students,],
                              "Total Budget": [total_budget] ,
                              "Average Math Score": [mean_math_score],
                              "Average Reading Score":[mean_reading_score],
                              "% Passing Math":[percentage_passing_math],
                              "% Passing Reading":[percentage_passing_reading],
                              "% Overall Passing Rate":[overall_passing_rate]})
#Format the Total Students and Budget
district_summary['Total Students']=district_summary['Total Students'].map('{:,}'.format)
district_summary['Total Budget']=district_summary['Total Budget'].map('${:,}'.format)

district_summary
```

```python
#SCHOOL SUMMARY
# Using GroupBy in order to separate the data into fields according to "school_name" values
grouped_schools_df= combined_lists.groupby(['school_name'])
grouped_schools_df.count()
```

```python
#Calculate school type
school_type=grouped_schools_df["type"].unique()
#Calculate the total number of students per school
total_students_school = combined_lists["school_name"].value_counts()
#Calculate the total school budget
total_school_budget=grouped_schools_df["budget"].unique()
#Calculate per student budget
per_student_budget=(total_school_budget)/(total_students_school)
#Calculate average math score
average_math_score=grouped_schools_df["math_score"].mean()
#Calculate average reading score
average_reading_score=grouped_schools_df["reading_score"].mean()
```

```python
#Calculate % passing math
pass_math=combined_lists[combined_lists["math_score"]>=70].groupby("school_name")
percent_pass_math=pass_math.count()["math_score"]/total_students_school*100
#Calculate %passing reading error
pass_read=combined_lists[combined_lists["reading_score"]>=70].groupby("school_name")
percent_pass_read=pass_read.count()["reading_score"]/total_students_school*100
#Calculate overall passing rate
overall_pass=(percent_pass_read + percent_pass_math)/2
                          
                          
                          
```

```python
# Place all of the data found into a summary DataFrame
school_summary = pd.DataFrame({"School Type": school_type, 
                               "Total Students": total_students_school,
                               "Total School Budget": total_school_budget,
                               "Per Student Budget":per_student_budget ,
                               "Average Math Score": average_math_score,
                               "Average Reading Score": average_reading_score,
                               "% Math Passing": percent_pass_math,
                               "% Reading Passing":percent_pass_read,
                               "% Overall Passing Rate": overall_pass })

school_summary
```

```python
# Sort and display top five schools in overall passing rate
high_pass = school_summary.sort_values("% Overall Passing Rate", ascending=False)
high_pass.head()
```

```python
# Sort and display five worst-performing schools in overall passing rate
low_pass = school_summary.sort_values("% Overall Passing Rate")
low_pass.head()


```

```python
#Math scores by grade
ninth = combined_lists[combined_lists["grade"]=="9th"]
tenth = combined_lists[combined_lists["grade"]=="10th"]
eleventh = combined_lists[combined_lists["grade"]=="11th"]
twelfth = combined_lists[combined_lists["grade"]=="12th"]
#find avg of math score for each grade
ninth_math_avg = ninth.groupby(["school_name"]).mean()["math_score"]
tenth_math_avg = tenth.groupby(["school_name"]).mean()["math_score"]
eleventh_math_avg = eleventh.groupby(["school_name"]).mean()["math_score"]
twelfth_math_avg = twelfth.groupby(["school_name"]).mean()["math_score"]
```

```python
#combine data to create dataframe
math_grade= pd.DataFrame({"9th": ninth_math_avg,
                                   "10th":tenth_math_avg,
                                   "11th":eleventh_math_avg,
                                   "12th":twelfth_math_avg})

math_grade
```

```python
#find avg of reading score for each grade
ninth_read_avg = ninth.groupby(["school_name"]).mean()["reading_score"]
tenth_read_avg = tenth.groupby(["school_name"]).mean()["reading_score"]
eleventh_read_avg = eleventh.groupby(["school_name"]).mean()["reading_score"]
twelfth_read_avg = twelfth.groupby(["school_name"]).mean()["reading_score"]
```

```python
#Scores by spending
spending_bins = [0, 585, 620, 645, 675]
group_names = ["<$585", "$585-619", "$620-644", "$645-675"]

school_summary["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels = group_names)

spending_math_score = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()['Average Math Score']
spending_reading_score = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()['Average Reading Score']
spending_passing_math =  school_summary.groupby(["Spending Ranges (Per Student)"]).mean()['% Math Passing']
spending_passing_reading =  school_summary.groupby(["Spending Ranges (Per Student)"]).mean()['% Reading Passing']
spending_overall_pass= school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing Rate"]



```

```python
#Create DataFrame
Spending_Score = pd.DataFrame({"Average Math Score":spending_math_score, "Average Reading Score":spending_reading_score,
                               "% Math Passing":spending_passing_math,"% Reading Passing":spending_passing_reading,
                                  "% Overall Passing Rate":spending_overall_pass}) 
Spending_Score
```

```python
# Scores by school size
size_bins = [0, 1000, 2000, 5000]
size_ranges = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_summary["School Size"] = pd.cut(school_summary["Total Students"], size_bins, labels = size_ranges)

avg_math_score = school_summary.groupby(["School Size"]).mean()["Average Math Score"]
avg_reading_score = school_summary.groupby(["School Size"]).mean()["Average Reading Score"]
percent_passing_math =  school_summary.groupby(["School Size"]).mean()["% Math Passing"]
percent_passing_reading =  school_summary.groupby(["School Size"]).mean()["% Reading Passing"]
overall_pass_rate = school_summary.groupby(["School Size"]).mean()["% Overall Passing Rate"]


# Create Dtaframe
Size_Score = pd.DataFrame({"Average Math Score":avg_math_score, "Average Reading Score":avg_reading_score,
                               "% Math Passing":percent_passing_math,"% Read Passing":percent_passing_reading,
                                    "% Overall Passing Rate":overall_pass_rate})   

Size_Score
```
